```python
#Importations
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


```

```python
#Reading the data and displaying the first 5 houses sold
df = pd.read_csv("ames.csv")
df.head()
```

```python
#Summary of the data-ames
df.info()
```

```python
df.describe()
```

```python
#Extracting SalePrice,TotRmsAbvGrd and OverallCond columns into a new data set(df1)
df1 = df[['SalePrice','TotRmsAbvGrd','OverallCond']]
df1.head()
```

```python
df1.info()
```

```python
df1.describe()
```

```python
# Calculate the mean,median and std of saleprice
mean=np.mean(df1['SalePrice'])
median=np.median(df1['SalePrice'])
Std=df1['SalePrice'].std()

# plotting one figure
plt.figure(figsize=(20, 14))
sns.set_style('darkgrid')
plt.subplot(3,2,1)


# visualizing the mean ,median and std of saleprice
# visualizing distribution of sale price
sns.histplot(data=df1, x='SalePrice', bins=50,color='brown', kde=True )
plt.title('Sales Distribution')
plt.axvline(mean,color='Green',label = 'Mean')
plt.axvline(median,color='blue',label = 'Median')
plt.axvline(mean-Std, color='red',linestyle='dashed', label = 'Lower quantile')
plt.axvline(mean+Std, color='purple',linestyle='dashed' ,label = 'Upper quantile')
plt.legend()


# visualizing saleprice distribution
# Sale price has a lot of outliers
plt.subplot(3,2,2)
sns.boxplot(y='SalePrice',data=df1 , color='brown')
plt.title('Distribution of SalesPrice')

# getting the mean,meadian and std of TotRmAbvGrd 
mean=np.mean(df1['TotRmsAbvGrd'])
median=np.median(df1['TotRmsAbvGrd'])
Std=df1['TotRmsAbvGrd'].std()

# Visualize TotRmAbvGrd distribution
plt.subplot(3,2,3)
sns.countplot(data=df1, x='TotRmsAbvGrd', color='green')
plt.title('Distribution of Total Rooms Above Grade')
plt.axvline(mean,color='Green',label = 'Mean')
plt.axvline(median,color='blue',label = 'Median')
plt.axvline(mean-Std, color='red',linestyle='dashed', label = 'Lower quantile')
plt.axvline(mean+Std, color='purple',linestyle='dashed' ,label = 'Upper quantile')
plt.legend()

plt.subplot(3,2,4)
sns.boxplot(y='TotRmsAbvGrd',data=df1 , color='green')
plt.title('Distribution of TotRmsAbvGrd')

# Getting the mean,median and std
mean=np.mean(df1['OverallCond'])
median=np.median(df1['OverallCond'])
Std=df1['OverallCond'].std()

# visualization of overall cond
plt.subplot(3,2,5)
sns.countplot(data=df1, x='OverallCond', color='b')
plt.title('Distribution of Overall Condition')
plt.axvline(mean,color='Green',label = 'Mean')
plt.axvline(median,color='blue',label = 'Median')
plt.axvline(mean-Std, color='red',linestyle='dashed', label = 'Lower quantile')
plt.axvline(mean+Std, color='purple',linestyle='dashed' ,label = 'Upper quantile')
plt.legend()

plt.subplot(3,2,6)
sns.boxplot(y='OverallCond',data=df1 , color='blue')
plt.title('Distribution of Overall Condition')

plt.show()
```

```python
# Overall condition with total saleprice per condition
Cond_sales = {}
Condition_sales =df1[['OverallCond','SalePrice']]
for condt in Condition_sales.OverallCond.unique():
    Cond_sales[condt] = [df1['SalePrice'][v] for v in df1[df1["OverallCond"] == condt].index]

Total_sales ={}
for key, value in Cond_sales.items():
    Total_sales[key] = sum(value)

sale_T = Total_sales.items()
Sorted_sales = sorted(sale_T, key=lambda x: x[0])
x = dict(Sorted_sales).keys()
y = dict(Sorted_sales).values()


# #plotting a bar plot for quantity against sales
fig, ax=plt.subplots(figsize=(10,6), ncols=2)

ax[0].bar(x,y)
ax[0].set_ylabel('Total salePrice')
ax[0].set_xlabel('Condition levels')
ax[0].set_title('SalePrice distribution per condition' )

ax[1].scatter(x,y)
ax[1].set_ylabel('Total salePrice')
ax[1].set_xlabel('Condition levels')
ax[1].set_title('SalePrice distribution per condition');

# highest sale price is made from houses under condition 5
# the least sale price is made from houses under condition 1,2,3,9 
```

```python
# subsets based on overall condition
below_average_condition=df[df['OverallCond']<5]
average_condition=df[df['OverallCond']==5]
above_average_condition = df[df['OverallCond']>5]

# use ggplot style
plt.style.use('ggplot')

# plot one figure
plt.figure(figsize=(18,6))
plt.subplot(2,3,1)

# Create custom bins so all are on the same scale
bins = range(df["SalePrice"].min(), df["SalePrice"].max(), int(df["SalePrice"].median()) // 20) # Plot three histograms, with reduced opacity (alpha) so 

# the three histograms will overlap
plt.hist( x=above_average_condition['SalePrice'],bins=bins, color='black',alpha=0.8,edgecolor='black')
plt.hist( x=average_condition['SalePrice'],bins=bins,color='green',alpha=0.3,edgecolor='green')
plt.hist( x=below_average_condition['SalePrice'],bins=bins, color='red',alpha=0.5,edgecolor='red')
plt.title('Distribution of saleprice grouped OverallCond')
plt.xlabel('SalePrice')
plt.ylabel('Number of Houses')

# set the legend 
plt.legend(['above_average_condition','average_condition','below_average_condition'])

# set the xticks
xticks = np.arange(50000, 800000, 50000)
plt.xticks(xticks, rotation=90)
plt.show()


# Most sales is made from average condition which is equal to 5
# least sales is made from below_average_condition which is below 5
```

```python
# plot one figure
plt.figure(figsize=(15, 8))

# subplot three shows how below average condition impacts saleprice
plt.subplot(2,3,1)
plt.hist( x=below_average_condition['SalePrice'],bins=bins, color='red',alpha=0.8,edgecolor='red')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.title('Below average condition')
plt.xticks(rotation=90)

# subplot two shows how average condition impacts saleprice
plt.subplot(2,3,2)
plt.hist( x=average_condition['SalePrice'],bins=bins,color='green',alpha=0.8,edgecolor='green')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.title('Average condition')
plt.xticks(rotation=90)

# subplot one shows how above average condition impacts saleprice
plt.subplot(2,3,3)
plt.hist( x=above_average_condition['SalePrice'],bins=bins, color='black',alpha=0.8,edgecolor='black')
plt.ylabel('No. of Houses')
plt.xlabel('SalePrice')
plt.xticks(rotation=90)
plt.title('Above average condition')

plt.show()

# Most saleprice is gotten from houses with average condition
# The amount of saleprice from the houses with below average condition is the least
```

```python
df2=df[['SalePrice','TotRmsAbvGrd','OverallCond']]
#Getting the best features that are highly negative and positive correlated to Saleprice  
Corr=df2.corr()

# heatmap to show correlation
sns.heatmap(Corr,cmap='coolwarm',vmin=-1,vmax=1,annot=True,square=True,
           annot_kws={'fontsize':11,'fontweight':'bold'})
plt.show()
#  TotRmsAbvGrd positively correlated to sale price
#  OverallCond negatively correlated to sale price.
```

```python
# creating a new feature age, the difference between Yrsold and YearBuilt
df['Age']=df['YrSold'] - df['YearBuilt']
df.head()
```

```python
 # The relationship btw age and saleprice
plt.figure(figsize=(16,7))
plt.subplot(2,2,1)
scat=sns.scatterplot(y='SalePrice',x='Age',data=df,legend=False,hue='OverallCond',palette='Set2',alpha=1.0)
scat.set(xlabel='Age of the House')
scat.set_ylabel('SalePrice')
scat.set(title='Relationship between saleprice and Age')
# most saleprice is made from the recent builthouses with age of 0-20years
# The saleprice of Houses of age btw 100-140 is the least

plt.subplot(2,2,2)
sns.boxplot(y='Age',data=df , color='green')
plt.title('Distribution of Age');

# The boxplot shows the median,maximum,minimum and the quartiles of age 
# From the scatter plot, house built recently make more sales than older houses
```
